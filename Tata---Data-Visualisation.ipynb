```python
## tata_data.xlsx

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tav= pd.read_excel('tata_data.xlsx')
tav.head()
```

```python
tav.info()
```

```python
tav.shape
```

```python
dup = tav.duplicated().sum()
dup
```

```python
tav = tav.drop_duplicates()
```

```python
tav['CustomerID'].isna().sum()
```

```python
infi = np.isinf(tav['CustomerID']) | np.isnan(tav['CustomerID'])
tav.loc[infi]
```

```python
#### replacing infinity and empty rows with zero
tav['CustomerID'] = tav['CustomerID'].fillna(0).replace(float('inf'),0)
```

```python
tav['CustomerID'] = tav['CustomerID'].astype('int')
```

```python
tav.describe()
```

```python
tav.columns
```

```python
tav[tav['CustomerID'] < 1]
```

```python
## removing rows with zero customerid

tav.drop(tav[tav['CustomerID']<1].index, inplace=True)
```

```python
## checking for rows with less than 1 quanttity

tav[tav['Quantity']<1]
```

```python
### remove the rows with less than 1 quantity

tav.drop(tav[tav['Quantity']<1].index, inplace =True)
```

```python
### checking for rows with zero unitprice

tav[tav['UnitPrice'] < 0.0001]
```

```python
tav.drop(tav[tav['UnitPrice']< 0.0001].index, inplace = True)
```

```python
tav['CustomerID'].unique()
```

```python
tav['CustomerID']= tav['CustomerID'].astype(str)
```

```python
tav.info()
```

```python
tav['Country'].unique()
```

```python
tav['Country'].nunique()
```

```python
tav[tav['Country'] == 'Unspecified']
```

```python
### removing rows with unspecified countries

tav.drop(tav[tav['Country'] == 'Unspecified'].index, inplace =True)
```

```python
### create new column Revenue by multiplying quantity and price

tav['Revenue'] = tav['Quantity'] * tav['UnitPrice']
tav['Revenue']
```

```python
### create new column profit by subtracting the price from revenue

tav['Profit'] = tav['Revenue'] - tav['UnitPrice'] 
tav['Profit']
```

```python
### creating a year and month column
import datetime as dt
tav['Year'] = tav['InvoiceDate'].dt.year
tav['Month'] = tav['InvoiceDate'].dt.strftime("%B")
tav
```

```python
### checking their date types

tav[['Year', 'Month']].info()
```

```python
### converting year column to date

tav['Year'] = pd.to_datetime(tav['Year'])
```

```python
tav[['Year', 'Month']].info()
```

```python
#### removing UK, AS IT IS NOT NEEDED FOR THE ANALYSIS

## droping UK FROM THE COUNTRY COLUMN

tav.drop(tav[tav['Country'] == 'United Kingdom'].index, inplace= True)
```

Top 10 Countries by Revenue (excluding UK)

```python
##3 creating a dataframe to show countries and revenue alone

tav1 = ['Country', 'Revenue']
tav1 = tav[tav1].copy()
tav1
```

```python
tav1.info()
```

```python
### to show top 10 revenue by country

rev = tav1.groupby(tav1['Country']).sum().sort_values('Revenue', ascending=False).head(10)
```

```python
rev
```

```python
### showing it using bar chart

sns.set_style("darkgrid")
rev.plot(kind = 'barh')
plt.title("top 10 revenue by country")
```

top 10 countries by units sold(excluding uk)

```python
## creating a dataframe to show countries and units sold alone

tav2 = ['Country', 'Quantity']
tav2 = tav[tav2].copy()
tav2
```

```python
### to show units sold per country

unit = tav2.groupby(tav2['Country']).sum().sort_values('Quantity', ascending=False).head(10)
```
